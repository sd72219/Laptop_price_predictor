```python
from sklearn.ensemble import AdaBoostRegressor
from sklearn.model_selection import train_test_split,GridSearchCV,RandomizedSearchCV
from sklearn.metrics import r2_score,mean_absolute_error,mean_squared_error
```

### Model Training:

```python
x = df.drop('Price',axis=1)
y = df['Price']
```

```python
x_train,x_test,y_train,y_test = train_test_split(x, y, test_size=0.2,random_state=1)
```

```python
Ada_reg = AdaBoostRegressor(random_state=1)
Ada_reg.fit(x_train,y_train)
```

```python
Ada_reg.score(x_train,y_train)
```

```python
Ada_reg.score(x_test,y_test)
```

```python
ada_reg = AdaBoostRegressor(random_state=1)

hyp = {'n_estimators':np.arange(10,100)}

Gscv_model = GridSearchCV(ada_reg,hyp,cv=5)
Gscv_model.fit(x_train,y_train)
```

```python
Gscv_model.best_estimator_
```

```python
ada_reg = AdaBoostRegressor(n_estimators=11,random_state=10)
ada_reg.fit(x_train,y_train)
```

```python
ada_reg.score(x_train,y_train)
```

```python
ada_reg.score(x_test,y_test)
```

### 2.Knn:

```python
from sklearn.neighbors import KNeighborsRegressor
from sklearn.preprocessing import StandardScaler,MinMaxScaler
```

```python
std_scale = StandardScaler()
x_scale1 = std_scale.fit_transform(x_train)
x_scale2 = std_scale.transform(x_test)
```

```python
x_df = pd.DataFrame(x_scale1,columns=x.columns)
x_df.head()
```

```python
x_dft = pd.DataFrame(x_scale2,columns=x.columns)
x_dft.head()
```

### Hyperparameter:

```python
knn_reg = KNeighborsRegressor()
param = {'n_neighbors':np.arange(2,10),
                   'p':[1,2]}
Gscv_model = GridSearchCV(knn_reg,param,cv=5)
Gscv_model.fit(x_df,y_train)
```

```python
Gscv_model.best_estimator_
```

```python
knn_model = KNeighborsRegressor(n_neighbors=3,p=1)
knn_model.fit(x_df,y_train)
```

```python
knn_model.score(x_dft,y_test)
```

```python
knn_model.score(x_df,y_train)
```

### Random Forest

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
ran_Reg = RandomForestRegressor(random_state=1)

hyp = {'n_estimators':np.arange(10,20),
       'max_depth':np.arange(5,10),
       'min_samples_split':np.arange(2,10),
       'min_samples_leaf':np.arange(1,10)}


Gscv_model2 = GridSearchCV(ran_Reg,hyp,cv=5)
Gscv_model2.fit(x_train,y_train)
```

```python
Gscv_model2.best_params_
```

```python
ran_Reg = RandomForestRegressor(n_estimators=18,
                                random_state=1,
                               max_depth=9,
                               min_samples_leaf=1,
                               min_samples_split=2)
ran_Reg.fit(x_train,y_train)
```

```python
ran_Reg.score(x_train,y_train)
```
